***Data Description: players.csv dataset***

196 observations, each row representing information about a player (196 players total)

```r
nrow(players)
```

7 variables indicating different information about each player

name:
- 196 chr values of players' names

hashedEmail
- 196 chr values of players' hashed email addresses

experience:
- 5 possible character values of Amateur, Beginner, Pro, Regular, Veteran indicating a player's experience level with the game
- distribution below

```r
experience_levels <- players |>
    group_by(experience) |>
    summarize(count = n())

experience_levels
```

subscribe:
- 2 possible lgl values of TRUE and FALSE indicating whether a player is subscribed to a game-related newsletter
- distribution (below) reveals that nearly 3/4 of the players are subscribed to a game-related newsletter, so it may be difficult to determine which factors predict being subscribed

```r
subscribed <- players |>
    group_by(subscribe) |>
    summarize(count = n())

subscribed
```

played_hours:
- dbl values indicating a player's total logged hours in the game
- see the lowest, highest, and average of hours played below
    - there seems to be very significant outliers, as the highest number of hours played is 223.1 hours, yet the average is only 5.8 hours

```r
hours <- players |>
    summarize(minimum = min(played_hours, na.rm = TRUE),
              maximum = max(played_hours, na.rm = TRUE),
              average = mean(played_hours, na.rm = TRUE))

hours
```

gender:
- 7 different chr values to indicate a player's (non)disclosed gender
- see distribution below

```r
gender <- players |>
    group_by(gender) |>
    summarize(count = n())

gender
```

age:
- 196 dbl values of players' ages
- see below for youngest, oldest, and average age of players

```r
age <- players |>
    summarize(minimum = min(Age, na.rm = TRUE),
              maximum = max(Age, na.rm = TRUE),
              average = mean(Age, na.rm = TRUE))

age
```

***Questions***

I will be addressing the broad question:
What player characteristics and behaviours are most predictive of subscribing to a game-related newsletter,
and how do these features differ between various player types?

The specific question I have formulated is: Can a player's number of hours played predict a player being subscribed to a game-related newsletter?

A greater number of hours played likely indicates more interested in a game, and therefore greater likelihood of subscribing to a game-related newsletter.

***Exploratory Data Analysis and Visualization***

- I have demonstrated in the code above that the datasets can be loaded into R
- The players.csv dataset, which I will be using, requires little wrangling as it is already tidy. Each column represents a variable, each row represents an observation, and each cell represents one value. However, I may be selecting only the relevant columns for my question. Further, any outliers may need to be removed
- The mean values for each quantitative variable in the players.csv data set can be found below

```r
players_mean <- players |>
    select(played_hours, Age) |>
    map_dfr(mean, na.rm = TRUE)

players_mean
```

Below are some exploratory visualizations of the data

```r
options(repr.plot.width = 14, repr.plot.height = 10)

hours_age <- players |>
    ggplot(aes(x = Age, y = played_hours, color = subscribe)) +
    geom_point() +
    labs(x = "Age", y = "Hours played", color = "Subscribed") +
    theme(text = element_text(size = 20))

hours_age
```

Given that most of the hours played are on the lower end (below 20), the few outliers of high amounts of hours played make it difficult to discern if there is a relationship between a player's age and the amount of hours they played

```r
hours_subscribed <- players |>
    ggplot(aes(x = played_hours, fill = subscribe)) +
    geom_histogram(color = "black") +
    labs(fill = "Subscribed to game-related newsletter?", x = "Hours played", y = "Number of players") +
    theme(text = element_text(size = 20))

hours_subscribed
```

This graph shows that, out of the few players in the dataset who are not subscribed to a game-related newsletter, most of them are on the lower end of the number of hours played. However, the imbalance between the great number of players with very low hours played and the few players with very high hours played make it difficult to gain further insight from this visualization.
